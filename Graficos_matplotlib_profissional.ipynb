```python
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rc('font', family = 'Arial') # Altera a fonte de todos os graficos em matplotlib
```

```python
meses = ['jan', 'fev', 'mar', 'abr', 'mai', 'jun', 'jul', 'ago', 'set', 'out', 'nov', 'dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]

plt.figure(figsize=(15, 8))
plt.plot(meses, vendas, marker = 'o')

plt.ylabel('Vendas (em milhares)')
plt.title('Crescimento das vendas ao Longo do Ano')

plt.show();
```

```python
# Grafico personalizado
meses = ['jan', 'fev', 'mar', 'abr', 'mai', 'jun', 'jul', 'ago', 'set', 'out', 'nov', 'dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]

plt.figure(figsize=(15, 8))
plt.plot(meses, vendas, marker = 'o')

# Adicionando um vspan para destacar o intervalo de meses com vendas a partir de julho
plt.axvspan(meses.index('jul'), meses.index('dez'), color='lightgreen', alpha=.15)

# Adicionar texto para destacar o intervalo de meses com vendas a partir de julho
plt.text(6, 80, 'Recupera√ß√£o de crescimento', fontsize=9, color='green', weight = 'bold')

plt.ylabel('Vendas (em milhares)')
plt.title('Crescimento das vendas ao Longo do Ano', weight='bold')

for posicao in range(len(meses)):
    plt.annotate(vendas[posicao], xy=(meses[posicao], vendas[posicao]),
                 xytext=(-5, 5), textcoords='offset points', ha='center', va='bottom',
                 color='green', weight='bold')
    
#plt.xticks([])
plt.tick_params(axis='x', length=0)
plt.box(False) # N√£o mostrar a caixa em volta do gr√°fico
plt.show();
```

# 2. **Gr√°fico de Barra**

```python
plt.figure(figsize=(8,6))
produtos = ['Produto A', 'Produto B', 'Produto C', 'Produto D', 'Produto E']
quantidades = [23, 45, 17, 30, 37]

barras = plt.bar(produtos, quantidades)

plt.title('Quantidades de Produtos Vendidos')
plt.show();
```

```python
plt.figure(figsize=(8,6))
produtos = ['Produto A', 'Produto B', 'Produto C', 'Produto D', 'Produto E']
quantidades = [23, 45, 17, 30, 37]

barras = plt.bar(produtos, quantidades)

# Valores nas barras
plt.bar_label(barras, labels=quantidades)

# retirando a borda do gr√°fico
plt.box(False)

# Retirando o eixo y
plt.yticks([])

plt.title('Quantidades de Produtos Vendidos')
plt.show();
```

```python
plt.figure(figsize=(8,6))
produtos = ['Produto A', 'Produto B', 'Produto C', 'Produto D', 'Produto E']
quantidades = [23, 45, 17, 30, 37]

barras = plt.bar(produtos, quantidades, color='lightgray')

# destacando o produto B
barras[1].set_color('teal')

# Valores nas barras
plt.bar_label(barras, labels=quantidades)

# retirando a borda do gr√°fico
plt.box(False)

# Retirando o eixo y
plt.yticks([])

plt.title('Quantidades de Produtos Vendidos')
plt.show();
```

```python
import pandas as pd

df = pd.DataFrame({'Produtos':['Produto A', 'Produto B', 'Produto C', 'Produto D', 'Produto E'], # Nome das frutas
                   'Quantidades':[23, 45, 17, 30, 37]})  # Quantidade vendida de cada fruta

df.sort_values(by='Quantidades',   # ordenar pela quantidade
               ascending=False,    # colocar na ordem contr√°ria a ascendente descendente <- maior para o menor)
               inplace=True,       # atualizar o DataFrame com as mudan√ßas
               ignore_index=True)  # ignorar √≠ndice anterior


plt.figure(figsize=(8,6))
barras = plt.bar(df['Produtos'], df['Quantidades'], color='lightgray')

# destacando o produto B
barras[0].set_color('teal')

# Valores nas barras
plt.bar_label(barras, labels=df['Quantidades'])

# retirando a borda do gr√°fico
plt.box(False)

# Retirando o eixo y
plt.yticks([])

plt.title('Quantidades de Produtos Vendidos', weight='bold')
plt.ylabel('Vendas (em milhares)')
plt.show();
```
